```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense,LeakyReLU,BatchNormalization,Flatten,Reshape
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

(x_train,y_train),(_,_)=mnist.load_data()
print(x_train)
x_train = x_train[y_train ==5]/255.0
x_train = np.expand_dims(x_train,axis=-1)
image_shape = (28,28,1)
latent_dim = 100
batch_size = 32
epochs = 1000
sample_interval = 10  
def build_discriminator():
    model = tf.keras.Sequential()
    model.add(Flatten(input_shape=image_shape))
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(1,activation='sigmoid'))
    model.summary()
    model.compile(loss='binary_crossentropy',
            optimizer=Adam(learning_rate=0.0002,beta_1=0.5),
            metrics=['accuracy'])
    return model            
def build_generator(latent_dim):
    model = tf.keras.Sequential()
    model.add(Dense(7*7*256,input_dim=latent_dim))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Reshape((7,7,256)))
    model.add(tf.keras.layers.Conv2DTranspose(128,(5,5),strides=(1,1),padding='same',activation='relu'))
    model.add(BatchNormalization(momentum=0.8))
    model.add(tf.keras.layers.Conv2DTranspose(64,(5,5),strides=(2,2),padding='same',activation='relu'))
    model.add(BatchNormalization(momentum=0.8))
    model.add(tf.keras.layers.Conv2DTranspose(1,(5,5),strides=(2,2),padding='same',activation='tanh'))
    model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.0002,beta_1=0.5))         
    return model
def build_gan(generator,discriminator):
    discriminator.trainable = True
    z = tf.keras.Input(shape=(latent_dim,))
    img = generator(z)
    valid = discriminator(img)
    gan_model = Model(z,valid)
    gan_model.compile(loss = 'binary_crossentropy',optimizer=Adam(learning_rate=0.0002,beta_1=0.5))
    return gan_model
discriminator = build_discriminator()
generator = build_generator(latent_dim)
gan_model = build_gan(generator,discriminator)
def train(epochs,batch_size,sample_interval):
    valid_y = np.ones((batch_size, 1))
    fake_y = np.zeros((batch_size,1))
    for epoch in range(epochs):
        idx = np.random.randint(0,x_train.shape[0],batch_size)
        imgs = x_train[idx]
        noise = np.random.normal(0,1,(batch_size,latent_dim))
        gen_imgs = generator.predict(noise)
        d_loss_real = discriminator.train_on_batch(imgs,valid_y)
        d_loss_fake = discriminator.train_on_batch(gen_imgs,fake_y)
        d_loss = 0.5 * np.add(d_loss_real,d_loss_fake)
        print("%d [D loss:%f]" %(epoch,d_loss[0]))
        discriminator.trainable = False
        noise = np.random.normal(0,1,(batch_size,latent_dim))
        valid_y = np.ones((batch_size,1))
        g_loss = gan_model.train_on_batch(noise,valid_y)
        print("%d [G loss:%f]" %(epoch,g_loss[0]))
        discriminator.trainable = True
        if epoch % sample_interval == 0:
            noise = np.random.normal(0,1,(1,latent_dim))
            gen_imgs = generator.predict(noise)
            plt.imshow(gen_imgs[0,:,:,0],cmap='gray')
            plt.show()
train(epochs, batch_size, sample_interval)
```

```python
import os
import glob
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Flatten
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Activation,AveragePooling2D,BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
import tensorflow as tf
# 定义一个函数get_train_data，用于获取训练数据
def get_train_data(directory, batch_size):
    train_datagen=ImageDataGenerator(rescale=1./255,  # 初始化ImageDataGenerator，用于数据增强
                                     shear_range=0.2,  # 设置剪切强度（逆时针方向的剪切变换角度）
                                     zoom_range=0.2,  # 设置随机缩放的幅度
                                     validation_split=0.2,  # 设置验证集的比例
                                     horizontal_flip=True)  # 随机水平翻转输入
    img_width,img_height =256,256  # 设置图像的宽度和高度
    train_generator =train_datagen.flow_from_directory(directory,  # 从文件夹中以实时数据增强的方式生成批数据
                                                       target_size=(img_width,img_height),  # 设置目标图像的尺寸
                                                       batch_size=batch_size)  # 设置批量大小
    validation_generator = train_datagen.flow_from_directory(directory,  # 从文件夹中以实时数据增强的方式生成批数据
                                                             target_size=(img_height, img_width),  # 设置目标图像的尺寸
                                                             batch_size=batch_size)  # 设置批量大小
    return train_generator, validation_generator  # 返回训练数据生成器和验证数据生成器
# 定义一个函数build_model，用于构建模型
def build_model(input_shape, num_classes):
    model = Sequential()  # 初始化一个神经网络模型
    model.add(Conv2D(32, (5, 5),input_shape=input_shape,activation='relu'))  # 添加一个卷积层，32个输出空间维度，5x5的窗口，输入形状为input_shape，激活函数为relu
    model.add(MaxPooling2D(pool_size=(3, 3)))  # 添加一个最大池化层，池化窗口大小为3x3
    model.add(Conv2D(64, (3, 3),activation='relu'))  # 添加一个卷积层，32个输出空间维度，3x3的窗口，激活函数为relu
    model.add(MaxPooling2D(pool_size=(2, 2)))  # 添加一个最大池化层，池化窗口大小为2x2
    model.add(Conv2D(128
                     , (3, 3),activation='relu'))  # 添加一个卷积层，64个输出空间维度，3x3的窗口，激活函数为relu
    model.add(MaxPooling2D(pool_size=(2, 2)))  # 添加一个最大池化层，池化窗口大小为2x2
    model.add(Flatten())  # 添加一个数据扁平化层
    model.add(Dense(512,activation='relu'))  # 添加一个全连接层，512个输出空间维度，激活函数为relu
    model.add(Dropout(0.25))  # 添加一个Dropout层，丢弃率为0.25
    model.add(Dense(128,activation='relu'))  # 添加一个全连接层，128个输出空间维度，激活函数为relu
    model.add(Dense(num_classes,activation='softmax'))  # 添加一个全连接层，num_classes个输出空间维度，激活函数为softmax
    model.summary()  # 打印模型的摘要信息
    return model  # 返回模型
# 定义一个函数train_model，用于训练模型
def train_model(model, train_generator, validation_generator, batch_size):
    opt=tf.keras.optimizers.Adam(learning_rate=0.001)  # 初始化Adam优化器，学习率为0.001
    model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])  # 编译模型，优化器为opt，损失函数为categorical_crossentropy，评价指标为accuracy
    train = model.fit(train_generator,  # 训练模型，训练数据为train_generator
                      steps_per_epoch=train_generator.samples // batch_size,  # 设置每个epoch的步数为样本总数除以批量大小
                      epochs=70,  # 设置epoch的数量为50
                      validation_data=validation_generator,  # 设置验证数据为validation_generator
                      validation_steps= validation_generator.samples // batch_size,  # 设置每个epoch的验证步数为样本总数除以批量大小
                      verbose=1)  # 设置日志显示模式为1（进度条）
    return model  # 返回模型
# 定义一个函数preprocess_image，用于预处理图像
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(256, 256))  # 加载图像，目标尺寸为256x256
    x = image.img_to_array(img)  # 将图像转换为数组
    x = x/255  # 将数组的元素值缩放到0~1之间
    return np.expand_dims(x, axis=0)  # 在数组的第0轴上增加一个维度
# 定义一个函数predict_image，用于预测图像
def predict_image(model, image_path):
    img = preprocess_image(image_path)  # 预处理图像
    result = model.predict(img)  # 使用模型预测图像
    y_pred_classes = np.argmax(result, axis=1)  # 获取预测结果中概率最大的类别的索引
    confidence = np.max(result)  # 获取预测结果中的最大概率值
    # 显示图像和预测结果
    img = img.squeeze()  # 去掉扩展的维度
    plt.imshow(img)  # 显示图像`
    plt.show()  # 显示图像和标题
    return y_pred_classes, confidence  # 返回预测的类别和概率值

train_dir = "小麦病虫害照片集/"  # 设置训练数据的目录
batch_size =256  # 设置批量大小
train_generator, validation_generator = get_train_data(train_dir, batch_size)  # 获取训练数据和验证数据
num_classes = len(glob.glob(train_dir+"/*"))  # 获取类别的数量
input_shape = (256, 256, 3)  # 设置输入的形状
model = build_model(input_shape, num_classes)  # 构建模型
model = train_model(model, train_generator, validation_generator, batch_size)  # 训练模型
model.save('wheat_disease_model.keras')  # 保存模型

file1="验证/00001293.jpg"  # 设置要预测的图像的路径
y_pred_classes, confidence = predict_image(model, file1)  # 预测图像
Classes = ['条锈病', '白粉病', '粘虫', '纹枯病', '蚜虫', '赤霉病', '麦蜘蛛']
print("Predicted class: {}, Confidence: {:.2f}".format(Classes[int(y_pred_classes)], confidence))
```
